```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ML Lab 3/canada_per_capita_income.csv')
df.head()
```

```python
df.insert(0, "bias", 1)
df.head()
```

```python
X = df.iloc[:, :-1].to_numpy()
Y = df['per capita income (US$)'].to_numpy()
```

```python
B = np.linalg.inv(X.T @ X) @ X.T @ Y
B
```

```python
y = B[0] + 2020 * B[1]
print("Per Capita Income in 2020: ", y)
```

```python
df = pd.read_csv('ML Lab 3\\salary.csv')
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df.insert(0, "bias", 1)
X = df.iloc[:, :-1].to_numpy()
Y = df['Salary'].to_numpy()
B = np.linalg.inv(X.T @ X) @ X.T @ Y
B
```

```python
y = B[0] + B[1] * 12
print("Salary at 12 years: ", y)
```

```python
df = pd.read_csv("ML Lab 3/hiring.csv")
df = df.replace({"experience": 
            {
                np.nan: 0,
                "two": 2,
                "three": 3,
                "five": 5,
                "seven": 7,
                "ten": 10,
                "eleven": 11
            }
            })
df["experience"] = df["experience"].astype(int)
df = df.dropna()
df
```

```python
df.insert(0, "bias", 1)
```

```python
X = df.iloc[:, :-1].to_numpy()
Y = df.iloc[:, -1].to_numpy()
B = np.linalg.inv(X.T @ X) @ X.T @ Y
B
```

```python
x = np.array([1, 2, 9, 6])
y = B @ x
print("1. ", y)
x = np.array([1, 12, 10, 10])
y = B @ x
print("2. ", y)
```

```python
df = pd.read_csv("ML Lab 3/1000_Companies.csv")
state_dummies = pd.get_dummies(df["State"], drop_first=True)
df = pd.concat([df.drop("State", axis=1), state_dummies], axis=1)
df = df.apply(pd.to_numeric)
bool_cols = df.select_dtypes(include="bool").columns
df[bool_cols] = df[bool_cols].astype(int)
df = df[[c for c in df.columns if c != "Profit"] + ["Profit"]]

df.head()
```

```python
df.insert(0, "bias", 1)
```

```python
X = df.iloc[:, :-1].to_numpy()
Y = df.iloc[:, -1].to_numpy()
B = np.linalg.inv(X.T @ X) @ X.T @ Y
B
```

```python
x = np.array([1, 91694.48, 515841.3, 11931.24, 1, 0])
y = B @ x
print("1. ", y)
```
